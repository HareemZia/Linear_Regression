```python
#GENERATING RANDOM DATA USING NP
import numpy as np
import matplotlib.pyplot as plt
seed=1
numInstance=200
np.random.seed(seed)
X=np.random.rand(numInstance,1).reshape(-1,1)
Y_true=-3*X+1
Y=Y_true+np.random.normal(size=(numInstance,1)).reshape(-1,1)
plt.scatter(X,Y,color='black')
plt.plot(X,Y_true,color='black', linewidth=3)
plt.title('True function Y=3X+1')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

#DATA SLICINIG AND MODEL FITTING
numTrain=20
numTest=numInstance-numTrain
X_Train=X[:-numTest]
X_Test=X[-numTest:]
Y_Train=Y[:-numTest]
Y_Test=Y[-numTest:]
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,r2_score
reg=linear_model.LinearRegression()
reg.fit(X_Train,Y_Train)
Y_Pred_test= reg.predict(X_Test)
Y_Pred_train=reg.predict(X_Train)
plt.scatter(Y_Test, Y_Pred_test,color='black')
plt.title("Plot of true values vs predicted values")
plt.xlabel("true values")
plt.ylabel("predicted values")
plt.show()

#MODEL VALIDATION
print("Root mean square error=%0.4f" %np.sqrt(mean_squared_error(Y_Test,Y_Pred_test)))
print("R_squared=%0.4f"%r2_score(Y_Test,Y_Pred_test))
print("Slope=",reg.coef_[0][0])
print("Intercept=",reg.intercept_[0])
plt.scatter(X_Test,Y_Test,color='Black')
plt.plot(X_Test,Y_Pred_test,color='black', linewidth=3)
titlestr="Predict Function Y=%0.2fX+%0.2f"%(reg.coef_[0],reg.intercept_[0])
plt.title(titlestr)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

#CORELATED VARIABLES
import numpy as np
seed=1
np.random.seed(seed)
X2=0.5*X
X2=0.5*X+np.random.normal(0, 0.04,size=numInstance).reshape(-1,1)
X3=0.5*X2+np.random.normal(0,0.04,size=numInstance).reshape(-1,1)
X4=0.5*X3+np.random.normal(0,0.04,size=numInstance).reshape(-1,1)
X5=0.5*X4+np.random.normal(0,0.04,size=numInstance).reshape(-1,1)

#CORELATIOM BETWEEN X AND X2
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(12, 9)) 
ax1.scatter(X, X2, color='black')
ax1.set_xlabel('X')
ax1.set_ylabel('X2')
C = np.corrcoef(np.column_stack((X[:-numTest], X2[:-numTest])).T)
title_str = 'Correlation between X and X2=%0.4f' % C[0, 1]
ax1.set_title(title_str)

ax2.scatter(X2, X3, color='black')
ax2.set_xlabel('X2')
ax2.set_ylabel('X3')
C2= np.corrcoef(np.column_stack((X2[:-numTest], X3[:-numTest])).T)
title_str2 = 'Correlation between X2 and X3=%0.4f' % C2[0, 1]
ax2.set_title(title_str2)

ax3.scatter(X3, X4, color='black')
ax3.set_xlabel('X3')
ax3.set_ylabel('X4')
C3= np.corrcoef(np.column_stack((X3[:-numTest], X4[:-numTest])).T)
title_str3 = 'Correlation between X3 and X4=%0.4f' % C3[0, 1]
ax3.set_title(title_str3)

ax4.scatter(X4, X5, color='black')
ax4.set_xlabel('X4')
ax4.set_ylabel('X5')
C4= np.corrcoef(np.column_stack((X4[:-numTest], X5[:-numTest])).T)
title_str4= 'Correlation between X4 and X5=%0.4f' % C4[0, 1]
ax4.set_title(title_str4)
plt.show()
```
